```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import os
```

```python
cwd = os.getcwd() 
lists = os.listdir(f'{cwd}/GWAS')
lists = [i.replace('.xlsx', '') for i in lists]
gwas = {}
for i in lists:
    gwas[i] = pd.read_excel(f'GWAS/{i}.xlsx')
```

```python
for k, v in gwas.items():
    gwas[k]['RSID'] = gwas[k]['RSID'].replace(',chr.+[AGTC]', '', regex = True)
    #gwas[k]['RSID'] = gwas[k]['RSID'].replace(';rs\d+', '', regex = True)
```

```python
gwases = pd.DataFrame()
for k, v in gwas.items():
    gwas[k]['Source'] = k
    gwases = pd.concat([gwases, gwas[k]])
    
```

# Анализ путей для разных комбинаций фенотипов, с целью выявления групп фенотипов, ассоциированные с которыми гены в ходе pathway анализа демонстрируют отчетливо значимый 1-2 метаболических пути.

```python
spisok = gwases.groupby('GENE_NAME').agg(pd.Series.tolist)[["RSID", "Source"]]
spisok['Count_snps'] = spisok.RSID.apply(lambda x: len(x))
spisok['Source'] = spisok.Source.apply(lambda x: set(x))
spisok['Count_source'] = spisok.Source.apply(lambda x: len(x))
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_seq_items', None)
pd.set_option('display.max_colwidth', 500)
pd.set_option('expand_frame_repr', True)
new_genes = spisok.sort_values(by='Count_source', ascending = False)[1::]
spisok.sort_values(by='Count_source', ascending = False)[1:15]
```

**Видим таблицу с наиболее представленными генами для разных групп фенотипов.   
Будем на нее опираться для поиска значимых метаболических путей.**   


```python
new_genes['Source_l'] = new_genes['Source'].apply(list)
```

```python
phenotypes = list(set(tuple(x) for x in list(new_genes.Source_l)))
```

## Осуществим поиск значимых snps с разными порогами отсечениями - менее и более строгим

**5e-8 - strict treshold - Pathway 2 in results    
1e-5 - Pathway 1 in results**

```python
for number in range(len(phenotypes)):
    k = list(phenotypes[number])
    unique_genes=[]
    for i in k:
        df = gwas[i]
        significant_rs = df[df['P'] < 1e-5 ]
        unique_gene = significant_rs['GENE_NAME'].unique()
        unique_gene = [gene for gene in unique_gene if gene != 'intergenic']
        for j in unique_gene:
            if j not in unique_genes:
                unique_genes.append(j)
    import gseapy as gp      
    try:
        enr_GOCC = gp.enrichr(gene_list=unique_genes, gene_sets=['GO_Cellular_Component_2021'], organism='Human', outdir=f'5 Results/Pathway 2/GOCC_{number}', cutoff=0.5)
        enr_Reactome = gp.enrichr(gene_list=unique_genes, gene_sets=['Reactome_2016'], organism='Human', outdir=f'5 Results/Pathway 2/Reactome_{number}', cutoff=0.5 )
        enr_GOMF = gp.enrichr(gene_list=unique_genes, gene_sets=['GO_Molecular_Function_2021'], organism='Human', outdir=f'5 Results/Pathway 2/GOMF_{number}', cutoff=0.5 )
        enr_GOBP = gp.enrichr(gene_list=unique_genes, gene_sets=['GO_Biological_Process_2021'], organism='Human', outdir=f'5 Results/Pathway 2/GOBP_{number}', cutoff=0.5 )
        enr_KEGG = gp.enrichr(gene_list=unique_genes, gene_sets=['KEGG_2021_Human'], organism='Human', outdir=f'5 Results/Pathway 2/KEGG_{number}', cutoff=0.5 )
    except:
        continue
```
