```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import os
```

```python
cwd = os.getcwd() 
lists = os.listdir(f'{cwd}/GWAS')
lists = [i.replace('.xlsx', '') for i in lists]
gwas = {}
for i in lists:
    gwas[i] = pd.read_excel(f'GWAS/{i}.xlsx')
```

```python
for k, v in gwas.items():
    gwas[k]['RSID'] = gwas[k]['RSID'].replace(',chr.+[AGTC]', '', regex = True)
    #gwas[k]['RSID'] = gwas[k]['RSID'].replace(';rs\d+', '', regex = True)
```

```python
gc = pd.read_excel('Add_files/Gene_corr.xlsx')
```

```python
gc
```

```python
data_cleaned = gc[['Фенотип 1', 'Фенотип 2', 'rg', 'SE']]
data_cleaned.dropna(inplace=True)
data_cleaned
```

```python
import numpy as np

phenotypes = pd.unique(data_cleaned[['Фенотип 1', 'Фенотип 2']].values.ravel('K'))

corr_matrix = pd.DataFrame(index=phenotypes, columns=phenotypes)
for i, row in data_cleaned.iterrows():
    pheno1 = row['Фенотип 1']
    pheno2 = row['Фенотип 2']
    rg_value = row['rg']
    corr_matrix.at[pheno1, pheno2] = rg_value
    corr_matrix.at[pheno2, pheno1] = rg_value

corr_matrix = corr_matrix.astype(float)
np.fill_diagonal(corr_matrix.values, 1)

corr_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))


sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, linewidths=0.5)

plt.title('Genetic Correlation Heatmap')
plt.xticks(rotation=90)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

```python
data_gwas_dsm = pd.DataFrame()
for k in ['DSM субклиническая депрессия logistic.top1000_new', 'DSM_БАР logistic.top1000_new',  'DSM_ГТР logistic.top1000_new', 'DSM_ДЕПРЕССИЯ logistic.top1000_new']  :
    data_gwas_dsm = pd.concat([gwases, gwas[k]])
    
```

```python
data_gwas_hads = pd.DataFrame()
for k in ['HADS_A. БАЛЛ linear.top1000_new', 'HADS_A_КАТЕГОРИЯ logistic.top1000_new', 'HADS_D. БАЛЛ linear.top1000_new', 'HADS_D_КАТЕГОРИЯ logistic.top1000_new', 'HADS_D_субклиническая депрессия logistic.top1000_new']:
    data_gwas_hads = pd.concat([gwases, gwas[k]])
```

```python


# Пример данных генетических корреляций (уже есть в corr_matrix)
corr_matrix = corr_matrix # Вставьте матрицу корреляций, созданную ранее

# Функция для вычисления чувствительности
def calculate_sensitivity(gwas_data):
    # Пример расчета чувствительности на основе статистики GWAS
    significant_hits = gwas_data[gwas_data['P'] < 1e-5]
    sensitivity = len(significant_hits) / len(gwas_data)
    return sensitivity

# Расчет чувствительности для DSM и HADS
sensitivity_dsm = calculate_sensitivity(data_gwas_dsm)
sensitivity_hads = calculate_sensitivity(data_gwas_hads)

print(f'Sensitivity DSM: {sensitivity_dsm}')
print(f'Sensitivity HADS: {sensitivity_hads}')
```

```python
import numpy as np

# Вычисление средней генетической корреляции между фенотипами DSM и HADS
def calculate_average_correlation(corr_matrix, group1, group2):
    correlations = []
    for pheno1 in group1:
        for pheno2 in group2:
            correlations.append(corr_matrix.at[pheno1, pheno2])
    return np.mean(correlations)


group_dsm =  ['DSM - генерализованное тревожное расстройство (ГТР)',
        'DSM субклиническая депрессия ']
group_hads = ['HADS-D субклиническая депрессия ', 'HADS-D депрессия ', 'HADS_A  (тревога) ', 'HADS_A  (тревога) общий балл КОЛИЧЕСТВЕННЫЙ ФЕНОТИП']


average_corr_dsm_hads = calculate_average_correlation(corr_matrix, group_dsm, group_hads)
print(f'Average Genetic Correlation between DSM and HADS: {average_corr_dsm_hads}')
```

```python
import matplotlib.pyplot as plt

# Визуализация чувствительности
labels = ['DSM', 'HADS']
sensitivities = [sensitivity_dsm, sensitivity_hads]

plt.bar(labels, sensitivities, color=['blue', 'orange'])
plt.xlabel('Phenotype Group')
plt.ylabel('Sensitivity')
plt.title('Sensitivity of DSM and HADS Phenotypes')
plt.show()

# Визуализация надежности (генетической корреляции)
plt.bar(['Average Genetic Correlation'], [average_corr_dsm_hads], color='green')
plt.xlabel('Correlation')
plt.ylabel('Value')
plt.title('Average Genetic Correlation between DSM and HADS Phenotypes')
plt.show()
```
